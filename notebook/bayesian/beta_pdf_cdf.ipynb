```python
import sys
import numpy as np
import scipy.special as ss
from scipy.stats import beta
from scipy import integrate
import matplotlib.pyplot as plt

def beta_pdf(x, a, b):
    if ss.beta(a, b) != 0:
        return x**(a-1)*(1-x)**(b-1)/ss.beta(a, b)
    else:
        return 0

def plot():
    x = np.arange(0, 1, .01)
    plt.figure(figsize=(12, 9))
    for (a, b) in [(.5, .5), (5, 1), (1, 3), (2, 2), (2, 5), (1, 1)]:
        plt.plot(x, beta_pdf(x, a, b), label='a={:.1f}, b={:.1f}'.format(a, b), lw=2)
    plt.legend(loc='upper center', frameon=False)
    plt.title('the pdf of beta function')
    plt.tick_params(direction='in')
    plt.ylim([0, 3.5])
    plt.ylabel('PDF')
    plt.savefig('./imgs/pdf.png')
    plt.show()

if __name__ == '__main__':
    plot()
```

```python
import sys
import numpy as np
from scipy.stats import beta
from scipy import integrate
import matplotlib.pyplot as plt

def plot():
    x = np.arange(0+0.0001, 1, .0001)
    plt.figure(figsize=(12, 9))
    for (a, b) in [(.5, .5), (5, 1), (1, 3), (2, 2), (2, 5), (1, 1)]:
        plt.plot(x, [(integrate.quad(beta_pdf, 0, x, args=(a, b)))[0] for x in x],
             label='a={:.1f}, b={:.1f}'.format(a, b), lw=2)
    plt.legend(loc='upper left', frameon=False)
    plt.ylabel('CDF')
    plt.title('the cdf of beta function')
    plt.tick_params(direction='in')
    plt.savefig('./imgs/pdf.png')
    plt.show()
    
if __name__ == '__main__':
    plot()
```
